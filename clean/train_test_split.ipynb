```python
import polars as pl
from sklearn.model_selection import train_test_split
```

```python
df = pl.read_parquet('/kaggle/input/nepbart-dataset-parquet/full_dataset.parquet')
```

```python
len(df)
```

```python
train, test = train_test_split(df, shuffle=True, test_size=0.1, random_state=34)
```

```python
del df
```

```python
len(train), len(test)
```

```python
train = train.unique(subset=['text'])
test = test.unique(subset=['text'])
```

```python
len(train), len(test)
```

```python
train.write_parquet('/kaggle/working/nepbart_train.parquet')
test.write_parquet('/kaggle/working/nepbart_test.parquet')
```

```python
del [train, test]
```
